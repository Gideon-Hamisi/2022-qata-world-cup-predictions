# Qatar2022-Football-World-Cup

```python
import pandas as pd
```

```python
histroy_matches = pd.read_csv('../input/qatar2022worldcupschudule/historical_win-loose-draw_ratios_qatar2022_teams.csv')
```

```python
matchs_schudule = pd.read_csv('../input/qatar2022worldcupschudule/matchs-schudule.csv',sep=';')
Qatar2022_teams = pd.read_csv('../input/qatar2022worldcupschudule/Qatar2022-teams.csv',sep=';')
```

# Data preprocessing

```python
histroy_matches
```

```python
matchs_schudule
```

```python
matchs_schudule = matchs_schudule.rename(columns = {'coutry2':'country2'})
```

```python
type(matchs_schudule['date'])
```

```python
matchs_schudule['date']=pd.to_datetime(matchs_schudule['date'],format='%d/%m/%Y')
```

```python
Qatar2022_teams.head()
```

```python
matchs_schudule['Team']=matchs_schudule['country1']
matches_groups = matchs_schudule.merge(Qatar2022_teams,on='Team').sort_values('date')
```

```python
matches_groups
```

```python
# get results for these teams
qatar_2022_teams_list = list(Qatar2022_teams['Team'])
qatar_2022_hist = histroy_matches.loc[(histroy_matches.country1.isin(qatar_2022_teams_list)) & (histroy_matches.country1.isin(qatar_2022_teams_list)),: ]

qatar_2022_hist
```

```python
qatar_2022_hist_2 = matches_groups.merge(qatar_2022_hist,on=['country1','country2'],how='inner').sort_values('date')
qatar_2022_hist_2
```

```python
country1_pred=[]
country2_pred=[]

for j in range(len(qatar_2022_hist_2)):
        if qatar_2022_hist_2['wins'][j]> qatar_2022_hist_2['looses'][j]:
            country1_pred.append(int(3))
            country2_pred.append(int(0))
        elif qatar_2022_hist_2['looses'][j]> qatar_2022_hist_2['wins'][j]:
            country1_pred.append(int(0))
            country2_pred.append(int(3))
        else:
            country1_pred.append(int(1))
            country2_pred.append(int(1))
                

```

```python
qatar_2022_hist_2.insert(11,'country1_pred',country1_pred)
qatar_2022_hist_2.insert(12,'country2_pred',country2_pred)
```

```python
qatar_2022_hist_2
```
